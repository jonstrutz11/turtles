# Timepoint Analysis with and without Wash

## Import Modules

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib as mpl

from turtles.turtles_utils import *
```

## Load Data

```python
#: Overall directory for TdT project data and analysis
tdt_dir = 'C:/Users/jonst/Box Sync/All_dNTP_data/'

#: Subdirectory for NGS run
data_dir = tdt_dir + 'Data_and_Results/All_Data_Filtered_Trimmed/Co_Washing_Controls_2/'

#: suffix of R1 fastq file to read in each directory
filename_end = 'trimmed.fq'

#: Directory to save generated figures in - it gets created if not present already
figure_dir = tdt_dir + 'Data_and_Results/Cobalt_Wash_Controls/Washing_Controls_Graphs_Final/'
os.makedirs(os.path.dirname(figure_dir), exist_ok=True)

#: Filenames are non-descriptive, so I created a table to map each filename to the relevant condition
id_file =tdt_dir + 'Data_and_Results/All_Data_Filtered_Trimmed/Washing_Controls_2_ids.csv'
```

```python
id_df = pd.read_csv(id_file, index_col='ID')
id_dict = id_df.to_dict('index')
```

## Load Sequences

```python
seqs_dict = read_seqs(data_dir, filename_end=filename_end, cutoff=5.8)
```

## Choose Legend Labels for Plotting

### Condition Labels

```python
condition_dict = {}
conditions = seqs_dict.keys()
for condition in conditions:
    cond = condition[:10]
    cond_name = id_dict[cond]['Condition_Name']
    condition_dict[condition] = cond_name

print(condition_dict)
```

### Replicate Labels

```python
rep_dict = {}
for condition in conditions:
    cond = condition[:10]
    rep_dict[condition] = id_dict[cond]['Replicate']
```

## Bin Sequences, Calculate dNTP Frequency, and Transform into Aitchison Space
Note: Binning can take a while.

```python
counts_dict = get_norm_len_base_counts(seqs_dict, num_bins=1000)
```

```python
pcts_dict = calc_norm_len_base_pcts(counts_dict)
```

```python
clr_data = calc_aitchison_distance(pcts_dict)
```

## Calculate and Plot Timepoint Results
### Convert Data into Long-Form pandas DataFrame
I know this looks messy, but it was difficult to just use generate_aitch_df as usual since I need to do so for many different zero and one conditions.

```python
data = pd.DataFrame()
averages_dict = {}

zero_pct_means = {}
for expt in ['Mg_NoWash_2uL', 'Mg_Wash_2uL', 'Mg_Wash_1uL', 'Mg_NoWash_2uL_3_ColRep']:

    if expt == 'Mg_NoWash_2uL':
        zero_control_conds = ['A3_FLD0193-236229090', 'A3_FLD0194-236239057', 'A3_FLD0195-236233060']
        one_control_conds = ['A3_FLD0199-236226073', 'A3_FLD0200-236232075', 'A3_FLD0201-236240034']
        expt_ids = [directory for directory in id_dict 
                    if id_dict[directory[:10]]['Wash_Condition'] == 'NoWash']
        expt_name = 'No Wash'
        zero_control_name = 'Mg_1hr'
        one_control_name = 'Co'

    elif expt == 'Mg_Wash_2uL':
        zero_control_conds = ['A3_FLD0209-236240039', 'A3_FLD0210-236236030', 'A3_FLD0211-236224109']
        one_control_conds = ['A3_FLD0215-236232081', 'A3_FLD0216-236227108']
        expt_ids = [directory for directory in id_dict 
                    if id_dict[directory[:10]]['Wash_Condition'] == 'Column'
                    and id_dict[directory[:10]]['Volume Ligated'] == 2]
        expt_name = 'Wash'
        zero_control_name = 'Mg_2uL'
        one_control_name = 'Co_2uL'
        
    elif expt == 'Mg_Wash_1uL':
        zero_control_conds = ['A3_FLD0217-236229098', 'A3_FLD0218-236245011', 'A3_FLD0219-236227109']
        one_control_conds = ['A3_FLD0223-236229101', 'A3_FLD0224-236244013']
        expt_ids = [directory for directory in id_dict 
                    if id_dict[directory[:10]]['Wash_Condition'] == 'Column'
                    and id_dict[directory[:10]]['Volume Ligated'] == 1]
        expt_name = None
        zero_control_name = 'Mg_1uL'
        one_control_name = 'Co_1uL'
        
    elif expt == 'Mg_NoWash_2uL_3_ColRep':
        zero_control_conds = ['A3_FLD0229-236238045', 'A3_FLD0230-236241043', 'A3_FLD0231-236229103']
        one_control_conds = zero_control_conds  # No one control
        expt_ids = [directory for directory in id_dict 
                    if id_dict[directory[:10]]['Wash_Condition'] == 'ColumnRep']
        expt_name = None
        zero_control_name = 'Mg3_Col'
        one_control_name = zero_control_name  # No one control

    zero_pct_means = {base: np.mean([clr_data[cond][base] for cond in zero_control_conds],
                                     axis=0) for base in ['A', 'C', 'G', 'T']}

    sub_pcts_dict = {key: pcts_dict[key] for key in pcts_dict if key[:10] in expt_ids}
    sub_clr_data = {key: clr_data[key] for key in clr_data if key[:10] in expt_ids}
        
    data_df = generate_aitch_df(sub_pcts_dict, sub_clr_data, condition_dict, rep_dict,
                                zero_control_conds, one_control_conds)

    data_df['Expt Condition'] = [expt_name] * len(data_df)
    
    n_bins = len(data_df['Bin Number'].unique())
    
    me_cond_col = []
    for directory in data_df.Directory.unique():
        me_cond_col += [id_dict[directory[:10]]['Me_Condition']] * n_bins
    
    data_df['Metal Condition'] = me_cond_col
    
    data_df = calc_signal(data_df, zero_control_conds, one_control_conds,
                          zero_control_name=zero_control_name,
                          one_control_name=one_control_name)
    
    data = pd.concat([data, data_df], axis=0, ignore_index=True)
    
    # For timepoint estimation
    if expt in ['Mg_NoWash_2uL', 'Mg_Wash_2uL']:
        len_dists = get_length_dists(seqs_dict)
        averages = parse_fastq_lengths(len_dists, expt_time=60)
        averages = calc_switch_bins(averages, data, mode='01')
        averages = calc_switch_times(averages, num_bins=1000, start_control_conds=zero_control_conds,
                                     end_control_conds=one_control_conds, t_expt=60)
    
        averages_dict[expt] = averages

data.head()
```

## Plot Results

### Set Seaborn Style

```python
sns.set(style='white', font='times-new-roman', font_scale=2)
```

### Plots for Mg
S13 A-D

```python
for base in ['A', 'C', 'G', 'T']:
    me_cond = 'Mg_0'
    sub_data = data.loc[data['Metal Condition'] == me_cond]
    sub_data = sub_data.loc[sub_data['Condition'] != 'No TdT']
    sub_data = sub_data.loc[sub_data['Condition'] != 'No TdT_2uL']

    length = 1000

    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,8))

    sns.lineplot(x='Bin Number', y='{} %'.format(base), hue='Expt Condition',
                 data=sub_data, ax=ax)

    ax.legend(frameon=False)

    ax.set_xlim(0, length)
    ax.set_xticks(range(0, length + 1, 200))
    ax.set_xticklabels(np.arange(0, 1.1, 0.2, dtype=np.float32), fontsize=16)

    ax.set_xlabel('Normalized Position')
    ax.set_ylabel('d{}TP Frequency'.format(base))
    ax.set_title('d{}TP Incorporation for Different Quenching Methods (Mg)'.format(base),
                 fontsize=24, pad=20)
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(labelleft=True, left=True, bottom=True)

    plt.tight_layout()
    fig.savefig(figure_dir + '{}%_Mg_Quench_Compare'.format(base))
    plt.show()
```

### Plots for Mg+Co
S13 E-H

```python
for base in ['A', 'C', 'G', 'T']:
    me_cond = 'Mg_0_Co_0'
    sub_data = data.loc[data['Metal Condition'] == me_cond]
    sub_data = sub_data.loc[sub_data['Condition'] != 'No TdT']
    sub_data = sub_data.loc[sub_data['Condition'] != 'No TdT_2uL']

    length = 1000

    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,8))

    sns.lineplot(x='Bin Number', y='{} %'.format(base), hue='Expt Condition',
                 data=sub_data, ax=ax)

    ax.legend(frameon=False)

    ax.set_xlim(0, length)
    ax.set_xticks(range(0, length + 1, 200))
    ax.set_xticklabels(np.arange(0, 1.1, 0.2, dtype=np.float32), fontsize=16)

    ax.set_xlabel('Normalized Position')
    ax.set_ylabel('d{}TP Frequency'.format(base))
    ax.set_title('d{}TP Incorporation for Different Quenching Methods (Mg+Co)'.format(base),
                 fontsize=24, pad=20)
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(labelleft=True, left=True, bottom=True)

    plt.tight_layout()
    fig.savefig(figure_dir + '{}%_Mg_Co_Quench_Compare'.format(base))
    plt.show()
```
